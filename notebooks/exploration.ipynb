# Terraform Security Misconfiguration – Exploratory Notebook

This notebook lets you inspect the dataset, features, and model predictions interactively.
Run `python main.py --train` first to generate the dataset and train the models.

```python
import sys, os
sys.path.insert(0, os.path.join(os.getcwd(), '..'))

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from src.feature_extractor import extract_features
from src.predictor import predict, print_prediction
```

## 1. Load the Dataset

```python
df = pd.read_csv('../data/terraform_dataset.csv')
print(f'Shape: {df.shape}')
df['risk_label'].value_counts()
```

## 2. Extract Features for the Full Dataset

```python
features_df = df['terraform_snippet'].apply(extract_features).apply(pd.Series)
features_df['risk_label'] = df['risk_label']
features_df.head()
```

## 3. Feature Correlation Heatmap

```python
numeric = features_df.drop(columns=['risk_label'])
plt.figure(figsize=(10, 7))
sns.heatmap(numeric.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()
```

## 4. Interactive Prediction on Custom Snippet

```python
my_snippet = '''
resource "aws_security_group" "test" {
  ingress {
    from_port   = 22
    to_port     = 22
    protocol    = "tcp"
    cidr_blocks = ["0.0.0.0/0"]
  }
}
'''

result = predict(my_snippet, model_name='RandomForest')
print_prediction(result)
```
